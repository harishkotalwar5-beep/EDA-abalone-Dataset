## 2. COMPLETE EXPLORATORY DATA ANALYSIS (EDA)


### 2.1 Outlier Detection using Box Plots


```python
numerical_cols = ["Length", "Diameter", "Height", "Whole weight", "Shucked weight", "Viscera weight", "Shell weight", "Rings"]
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
axes = axes.flatten()
for i, col in enumerate(numerical_cols):
    sns.boxplot(y=abalone[col], ax=axes[i], color="lightblue")
    axes[i].set_title(f"Boxplot of {col}")
    axes[i].set_ylabel(col)
plt.suptitle("Outlier Detection - Box Plots", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
print("Height outliers detected (max=1.13 vs Q3~0.165) - contains extreme values!")
```

### 2.2 Distribution Analysis - Histograms of All Numerical Features


```python
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
axes = axes.flatten()
for i, col in enumerate(numerical_cols):
    axes[i].hist(abalone[col], bins=30, color="steelblue", edgecolor="black", alpha=0.7)
    axes[i].set_title(f"Distribution of {col}")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Frequency")
plt.suptitle("Distribution of All Numerical Features", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

### 2.3 Pairplot - Relationships Between All Numerical Features


```python
# Pairplot (using a sample for speed)
sample_df = abalone.sample(n=500, random_state=42)
sns.pairplot(sample_df[numerical_cols], diag_kind="kde", plot_kws={"alpha": 0.4})
plt.suptitle("Pairplot of Numerical Features (500-sample)", y=1.02, fontsize=14)
plt.show()
```

### 2.4 Feature vs Target (Rings) - Scatter Plots for All Features


```python
features = ["Length", "Diameter", "Height", "Whole weight", "Shucked weight", "Viscera weight", "Shell weight"]
fig, axes = plt.subplots(2, 4, figsize=(18, 8))
axes = axes.flatten()
for i, col in enumerate(features):
    axes[i].scatter(abalone[col], abalone["Rings"], alpha=0.3, color="royalblue", s=5)
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Rings")
    axes[i].set_title(f"{col} vs Rings")
# Hide the last unused subplot
axes[-1].set_visible(False)
plt.suptitle("Feature vs Target (Rings) Scatter Plots", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

### 2.5 Violin Plots - Feature Distribution by Sex Category


```python
# After encoding: 0=Male, 1=Female, 2=Infant
sex_labels = {0: "Male", 1: "Female", 2: "Infant"}
abalone["Sex_label"] = abalone["Sex"].map(sex_labels)

fig, axes = plt.subplots(2, 4, figsize=(18, 8))
axes = axes.flatten()
for i, col in enumerate(features):
    sns.violinplot(x="Sex_label", y=col, data=abalone, ax=axes[i], palette="Set2")
    axes[i].set_title(f"{col} by Sex")
axes[-1].set_visible(False)
plt.suptitle("Feature Distribution by Sex (Violin Plots)", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

### 2.7 Statistical Summary & Skewness Analysis

### 2.8 Top Correlations with Target Variable (Rings)


```python
corr_with_rings = abalone[numerical_cols].corr()["Rings"].drop("Rings").sort_values(ascending=False)
plt.figure(figsize=(8, 5))
corr_with_rings.plot(kind="bar", color=["green" if x > 0 else "red" for x in corr_with_rings], edgecolor="black")
plt.title("Correlation of Features with Target Variable (Rings)")
plt.xlabel("Feature")
plt.ylabel("Pearson Correlation")
plt.xticks(rotation=45)
plt.axhline(0, color="black", linewidth=0.8)
plt.tight_layout()
plt.show()
print(corr_with_rings)
```

### 2.9 Age Group Analysis


```python
# Bin rings into age groups
bins = [0, 7, 10, 15, 30]
labels = ["Young (1-7)", "Middle (8-10)", "Mature (11-15)", "Old (16+)"]
abalone["Age_Group"] = pd.cut(abalone["Rings"], bins=bins, labels=labels)

plt.figure(figsize=(8, 5))
abalone["Age_Group"].value_counts().sort_index().plot(kind="bar", color="coral", edgecolor="black")
plt.title("Distribution of Abalone Age Groups")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

print("Age Group Distribution:")
print(abalone["Age_Group"].value_counts().sort_index())
```

### 2.10 Feature Importance (via Random Forest)


```python
from sklearn.ensemble import RandomForestRegressor as RFR
import numpy as np

feature_names = ["Sex", "Length", "Diameter", "Height", "Whole weight", "Shucked weight", "Viscera weight", "Shell weight"]
X_feat = abalone.drop(columns=["Rings", "Sex_label", "Age_Group"], errors="ignore")
y_feat = abalone["Rings"]

rf_model = RFR(n_estimators=100, random_state=42)
rf_model.fit(X_feat, y_feat)

importances = rf_model.feature_importances_
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(10, 5))
plt.bar(range(len(feature_names)), importances[indices], align="center", color="teal", edgecolor="black")
plt.xticks(range(len(feature_names)), [feature_names[i] for i in indices], rotation=45)
plt.title("Feature Importance (Random Forest)")
plt.xlabel("Feature")
plt.ylabel("Importance Score")
plt.tight_layout()
plt.show()

print("Feature Importances:")
for i in indices:
    print(f"  {feature_names[i]}: {importances[i]:.4f}")
```

SPLITING DATA SET


```python
from sklearn.model_selection import train_test_split
X = abalone.drop('Rings',axis=1)
Y = abalone['Rings']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

FEATURE SCALLING
(standardization bring all feature to similar scale which is crucial for gradient based models like linear regression)


```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

APPLYING MANY REGRESSION MODEL



```python
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

models = [
    ('Linear Regression', LinearRegression()),
    ('Ridge Regression', Ridge()),
    ('Decision Tree', DecisionTreeRegressor(random_state=42)),
    ('Random Forest', RandomForestRegressor(random_state=42))
]

results = {}
for name, model in models:
    model.fit(X_train, Y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(Y_test, y_pred)
    r2 = r2_score(Y_test, y_pred)
    results[name] = r2
    print(f'{name}: MSE = {mse:.2f}, R2 = {r2:.2f}')
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=list(results.keys()), y=list(results.values()))
plt.title('Comaprision of regression models(R2 score)')
plt.xlabel('Model')
plt.ylabel('R2 Score')
plt.xticks(rotation=30)
plt.show()
```

```python
dtr = DecisionTreeRegressor()
dtr.fit(X_train,Y_train)
y_pred = dtr.predict(X_test)
print("Decison Tree MSE:", mean_squared_error(Y_test, y_pred))
print("Decision Tree R2 Score:", r2_score(Y_test, y_pred))
```

```python
rdm = RandomForestRegressor()
rdm.fit(X_train,Y_train)
y_pred = rdm.predict(X_test)
print("Random Forest MSE:", mean_squared_error(Y_test, y_pred))
print("Random Forest R2 Score:", r2_score(Y_test, y_pred))
```
